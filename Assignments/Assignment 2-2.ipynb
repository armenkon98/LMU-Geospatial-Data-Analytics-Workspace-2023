In this problem you should:

```python
gb = kumpula_may.groupby('YR--MODAHRMN')
```

```python
kumpula_may = kumpula_may.set_index('YR--MODAHRMN') #set index
kumpula_may.index = pd.to_datetime(kumpula_may.index, format='%Y%m%d%H%M') #index to datetime
```

```python
# this cell splits the dataframe into a list with multiple dataframes based on day
df = kumpula_may.copy()
df = df.reset_index()
splits= list(df.groupby(df['YR--MODAHRMN'].dt.day))
splits
```

```python
mean_value = []
max_value = []
min_value = []
kumpula_may_daily = []
for df in splits:
    mean_value = df[1]['Celsius'].mean() 
    max_value = df[1]['Celsius'].max() 
    min_value = df[1]['Celsius'].min() 
    row1 = [mean_value,max_value,min_value]
    kumpula_may_daily.append(row1)
    
kumpula_may_daily = pd.DataFrame(kumpula_may_daily, columns = ['mean', 'max', 'min'])
kumpula_may_daily
```

```python
dates =[] #get date index from split
for df in splits:
    df = df[1].reset_index()
    date_at_each_day = df.loc[0,'YR--MODAHRMN']
    dates.append(date_at_each_day)

kumpula_may_daily.index = dates
```
